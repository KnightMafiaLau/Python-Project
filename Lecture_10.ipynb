```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

raw_df = pd.read_csv('winter_olympics.csv')

print (raw_df['Country'])
```

```python
Norway_df = raw_df.loc[raw_df['Country'] == 'Norway']
Norway_df1 = Norway_df
ax = Norway_df1.groupby('Sport').Medal.count().plot(kind = 'barh').set_xlabel('Norway Medal Count')
```

Activity 2: Create a grouped or stacked bar chart that distinguishes gold, silver, and bronze medals for the country selected in Activity 1. (10 points)

```python
Norway_df2 = Norway_df
ax2 = Norway_df2.groupby('Sport').Medal.value_counts().plot(kind = 'barh').set_xlabel('Norway Medal Count')
```

Activity 3: Choose any three sports. Using Matplotlib, visualize the count of medals awarded by year for each of those sports. As an example, the following image uses a side-by-side grouped bar chart to count medals for skiing, bobsled, and figure skating. (10 points)

```python
sport_df = raw_df.loc[(raw_df['Sport'] == 'Alpine Skiing') | (raw_df['Sport'] == 'Bobsled') | (raw_df['Sport'] == 'Figure Skating')]
ax3 = sport_df.groupby('Sport').Year.value_counts().unstack(0).plot(kind = 'bar').set_ylabel('Medal Count')
plt.savefig('Medal_count.png')
```
