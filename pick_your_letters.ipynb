```python
import random
import string
```

```python
def read_from_file(file_name):
    """
    Reads all words from file
    Parameter file_name is the name of the file
    This function returns a list containing all the words
    """
    f = open(file_name, "r")
    all_words = f.read().splitlines()
    f.close()
    return all_words
```

```python
def ask_for_length():
    #loop until the input is correct
    while True:
      print ('Enter a number between 3 - 10 to be length of the word you are going to guess:')

      l = input()
      if l.isnumeric():
        #when l is a number, we change it to integer type
        l = int(l)
        if 3 <= l <= 10:
          #if the number is in the range, then return
          return l
        else:
          print('The number should be between 3 - 10.')
      
      else:
        #when the input is not a number
        print ('Please enter a number between 3 - 10.')
```

```python
def filter_word_list(all_words, length):
    word_list = [x for x in all_words if len(x) == length]
    return word_list
```

```python
def set_up(length):
  alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u',\
'v', 'w', 'x', 'y', 'z']
  main_pile = alphabet * length
  discard_pile = []
  return (main_pile, discard_pile)
```

```python
def shuffle_cards(pile):
  random.shuffle(pile)
    # TODO
```

```python
def move_to_discard_pile(discard_pile, card):
  discard_pile.append(card)

```

```python
def deal_initial_cards(main_pile, discard_pile, length):
    #initialize the hand card
    computer_card = []
    player_card = []
    #main deal loop
    for i in range(length):
      #give one card to computer, and drop from the main pile
      computer_card.append(main_pile[-1])
      main_pile.pop()
      player_card.append(main_pile[-1])
      main_pile.pop()
    
    discard_pile.append(main_pile[-1])
    main_pile.pop()

    return (player_card, computer_card)
 
```

```python
def get_first_from_pile_and_remove(pile):
    return pile.pop()
```

```python
def check_bricks(main_pile, discard_pile):
    if len(main_pile) > 0:
 
      return
    else:
      shuffle_cards(discard_pile)
      main_pile = discard_pile
      discard_pile = []
      discard_pile.append(main_pile[-1])
      main_pile.pop()
```

```python
def computer_play(computer_hand_cards, computer_target_list, main_pile, discard_pile, filtered_word_list):
  #get the first card on the discard_pile
  hold_card = discard_pile[-1]

  #if the hold card can increase the max similarity, then computer take it
  dit = target_list(computer_hand_cards, filtered_word_list)
  current_max = dit[max(dit)]
  for i, j in enumerate(computer_hand_cards):
    temp_hand_cards = computer_hand_cards.copy() 
    temp_hand_cards[i] = hold_card
    temp_dit = target_list(temp_hand_cards, filtered_word_list)
    temp_max = temp_dit[max(temp_dit)]

    if temp_max > current_max:
      print ('Computer took\'',hold_card,'\' from DISCARD PILE')
      computer_hand_cards[i] = hold_card
      get_first_from_pile_and_remove(discard_pile)
      return
    
  #when not pick the card from discard pile, reveal the letter from main pile and make the decision

  hold_card = get_first_from_pile_and_remove(main_pile)
  move_to_discard_pile(discard_pile, hold_card)

  for i, j in enumerate(computer_hand_cards):
    temp_hand_cards = computer_hand_cards.copy() 
    temp_hand_cards[i] = hold_card
    temp_dit = target_list(temp_hand_cards, filtered_word_list)
    temp_max = temp_dit[max(temp_dit)]

    if temp_max > current_max:
      print ('Computer took\'',hold_card,'\' from MAIN PILE')
      computer_hand_cards[i] = hold_card
      get_first_from_pile_and_remove(discard_pile)
      return

  print ('Computer discard\'',hold_card,'\' from MAIN PILE')
```

```python
def target_list(computer_hand_cards, filtered_word_list):
  # basic logic is to search all words similarity
  #to the current hand cards and make a dict by their similarity


  #first, make the current hand card list into string
  computer_words = ''
  for ele in computer_hand_cards:
    computer_words += ele
  #I use the filtered list as the target list
  #compare the filtered list with the current hand card, get how many similar letters for each word as dit
  #we only want those words with max similarity or 1 less than the max similarity
  dit = {'a':1}
  max_similarity = 0 #store the current max similarity letters count
  
  for i, j in enumerate(filtered_word_list):
    similarity = sum(a == b for a, b in zip(computer_words, j))

    max_similarity = dit[max(dit)]
    if similarity > max_similarity:
      dit = {k:v for (k, v) in dit.items() if v + 1 >= similarity}
      dit[j] = similarity
    elif similarity + 1 >= max_similarity:
      dit[j] = similarity
  

  return dit


 
```

```python
def ask_for_the_letter_to_be_replaced(length):
    print('Input the index of the letter to be replaced, e.g. \'1\':')
    while True:
      pos = input()
      if pos.isnumeric() and length >= int(pos) > 0:
        return pos
      
      print ('Please input a valid number')
    
```
